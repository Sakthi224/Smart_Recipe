```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/indian_food.csv")
```

```python
df = df.dropna(subset=['region'])
```

```python
df['region'] = df['region'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
def clean_ingredients(ing_str):
    return set(ing_str.lower().replace(',', '').split())
```

```python
df['ingredients_set'] = df['ingredients'].apply(clean_ingredients)
```

```python
df[['name', 'ingredients_set']].head()
```

```python
user_input = input("Enter ingredients you have (comma-separated):\n")
```

```python
user_ingredients = set(user_input.lower().replace(',', '').split())
```

```python
print(user_ingredients)
```

```python
def jaccard(set1, set2):
    inter = set1 & set2
    union = set1 | set2
    return len(inter) / len(union) if union else 0
```

```python
jaccard({'onion', 'tomato'}, {'onion', 'potato'})
```

```python
df['similarity'] = df['ingredients_set'].apply(lambda x: jaccard(user_ingredients, x))
```

```python
df[['name', 'similarity']].sort_values(by='similarity', ascending=False).head()
```

```python
df_sorted = df.sort_values(by='similarity', ascending=False).head(5)
```

```python
for _, row in df_sorted.iterrows():
    print("\n Recipe:", row['name'])
    print(" Ingredients:", row['ingredients'])
    print(" Diet:", row['diet'])
    print(" Prep Time:", row['prep_time'], "mins")
    print(" Cook Time:", row['cook_time'], "mins")
    print(" Flavor Profile:", row['flavor_profile'])
    print(" Course:", row['course'])
    print(" Region:", row['region'])
```
